```python
## barplot to show the average age across status group

# status group = age > 30,  status   = single

sns.barplot(data = df, x = "status_group", y = "age", ci = None)

plt.title('age Statistic by status_group')
plt.grid()
```

```python
# create a barplot showing the average income by good religious across smokes 
plt.figure(figsize=(20,7))

sns.barplot(data = df, x = "good_religion", y = "income", hue = "smokes", ci = None)

plt.xticks(rotation = -45)
plt.title('income Statistic by good religion across smokes')
plt.legend(loc = 'upper left')

#good religious = likes (dogs and cats), income > 20K
```

```python
df['drugs'].replace(['never', 'sometimes', 'often', 'unknown'], [0, 1, 2, 3], inplace = True)
df.head()
```
